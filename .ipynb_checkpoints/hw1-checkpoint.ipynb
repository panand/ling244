```python
#!/usr/bin/env python

'''Creates a PigLatin generator.'''

__author__ = 'ajschwade'
__email__ = "aschwade@ucsc.edu"
__credits__ = ["Allan Jay Schwade"]

def pigLatin(w):
    """
	this function handles pigLatin for word w

    Args:
		w: a word

    Returns:
        pigLatined: a PigLatin representation of word w

    Raises:
        Nothing

	"""
    #These arrays specify which characters count as vowels and consonants.
    vowels=["a","e","i","o","u"] 
    consonants=["q","w","r","t","y","p","s","d","f","g","h","j","k","l","z","x","c","v","b","n","m"]
    
    ### Filters ###
    #If the word contain illicit characters, the function will not pig it.
    for ch in list(w):        
        if (ch not in vowels) and (ch not in consonants):
            return w
        else:
            pass
        
    #If the pig-ending would be affixed to a vowel, the final vowel is removed.
    if (w[0] in vowels) and (w[-1] in vowels):
        w=w[0:-1]
    

    ### The Main Function ###
    #This for-loop looks at each character for a given word (starting with the first) and pops it to the end of the array until it finds a vowel,
    # in which case it halts and returns to resulting string plus the pig ending.
    for ch in list(w):
        if ch not in vowels:
            w=w[1:]+w[0]
        else:
            return w +"ay"
        
        
def pigLatinParts(text):
    """
	this function handles pigLatin for all words in string S. It also remembers the punctuation as at the last Goedel number.

    Args:
		text: a string

    Returns:
        pigLatined: what people can download and use (a program)
        
    Raises:
        Nothing

	"""
    #Defines which characters count as punctuation and creates an array to hold the pigged sentence.
    punc=[ '.' , ',' ]
    pigSentence=[]
    
    #This loop splits the input string into an array containing the words of the input sentence and pigs the words in that list 
    for word in text.split(' '):
        # If a word in the array contains some kind of puntuation, the punctuation is removed from 
        # the element, the word is pigged, and then the punctuation is re-attached. Else, the word is simply
        # pigged. The pigged array is the then printed as a string.
        if word[-1] in punc:
            hold=word[-1]
            word=word[0:-1]
            pigWord=pigLatin(word)+(hold)
            pigSentence.append(pigWord)
        else:
            pigWord=pigLatin(word)
            pigSentence.append(pigWord)
            
    return ' '.join(pigSentence)

if __name__ == "__main__":
    
    # all of the below code is used to TEST whether yours worked
    
    inputLine = "if you think, sadly, that he will actually visit, then I am fine."
    key = "ifay ouyay inkthay, adlysay, atthay ehay illway actuallyay isitvay, enthay I amay inefay." 
    pg = pigLatinParts(inputLine)
    
    if pg==key:
        result = "PASS"
    else:
        result = "FAIL"
    print inputLine +'\n' + pg +'\n' + result 
```

```python
#!/usr/bin/python2.7

#Imports various NLTK modules. The wordnet corpus is used to retrieve word definitions and their part of speech. The Carnegie Mellon 
#University Pronunciation Dictionary (CMUPD) is used to retrive the IPA transcriptions of words.
from nltk.corpus import wordnet as wn
import nltk.corpus.reader.cmudict as cmu

#Specifies the directory in which the CMUPD can be found
cmucorp=cmu.CMUDictCorpusReader('/Users/allanschwade/Google Drive/Academia/Documents/UCSC/Computational Linguistics/ling244/','cmudict7a.txt')

#This key will be used to convert the CMUPD's pronunciation into standard IPA characters
cmu_ipa_key={'AA':'ɑ','AH':'ʌ','AO':'ɔ','AW':'aʊ','AY':'aɪ','B':'b','CH':'tʃ','D':'d','DH':'ð','EH':'ɛ','ER':'ɚ','EY':'eɪ','F':'f','G':'g','HH':'h','IH':'ɪ','IY':'i','JH':'dʒ','K':'k','L':'l','M':'m','N':'n','NG':'ŋ','OW':'oʊ','OY':u'ɔɪ','P':'p','R':'ɹ','S':'s','SH':'ʃ','T':'t','TH':'θ','UH':'ʊ','UW':'','V':'v','W':'w','Y':'j','Z':'z','ZH':'ʒ'}

#Text containing words that need to be turned into a dictionary
string = '''When I was little, I thought a python was a snake.
            Now I know it is really a monster in disguise.
            But I will prevail over this hideous beast.'''

#Coverts the string to lowercase then changes the objects type to a list, wherein each element is a word
lcase=string.lower()
list_of_words=lcase.split()

#Since the text contains punctuation which is undesirable for dictionary entries the loop below
#looks to see which words end with punctuation and removes them from the words.
punc=[ '.' , ',' ]
unpuncd=[]
for word in list_of_words:
    if word[-1] in punc:
        word=word[0:-1]
        unpuncd.append(word)
    else:
        unpuncd.append(word)
        
#Duplicate words are removed from the list.
final_list=list(set(unpuncd))

### The Dictionary ###
#Creates a dictionary to store the words in. 
dictionary = {}

#This loop retreives the definition and part of speech from the wordnet corpus and the pronunciation of 
#the word from the CMUPD and stores the values in a dictionary.
for word in final_list:
    # In order to retrieve definitions from wordnet you must supply the word, part of speech, and
    # entry number. The key takes the form "word.pos.#".  Since the words in "final list" have
    # different parts of speech, in order to automate the lookup process, the following loop plugs in
    # the different parts of speech one at a time into the look up key. If a word has no entry for the given
    # part of speech the relevant values are set to 'NA' and the functions continues until it sucessfully 
    # retrives a definition.
    for pos in ['r','n','v','a']:
        try:
            lookUpKey= word+'.'+pos+'.01'
            definition=wn.synset(lookUpKey).definition
            #pos=pos
            break
        except:
            definition="NA"
            pos="NA"
            pass
        
    #The pronunciations for my dictionary are retirved from the CMUPD. This function looks into the
    #corpus file earlier, organizes it into a dictionary, and using the word as the key, retrives the definition.
    raw_phon=cmucorp.dict()[word]
    phon=""
    
    #Segments in the CMUPD are marked to show stress using a number. Since numbers are not
    #desireable in our dictionary pronunciations, this function removes the numbers from the segments.
    for seg in raw_phon[0]:
        if seg[-1] in "012":
            char=seg[0:-1]
            phon=phon+cmu_ipa_key[char]
        else:
            phon=phon+cmu_ipa_key[seg]
    
    #The word definitions, part of speech, and pronunciation are finally placed into an entry.
    dictionary[word]={'def':definition, 'pos':pos, 'phon':phon}

### Corrections ###
# Since wordnet does not have a dictionary for all the words in question, some entries must be 
# manually entered. This function streamlines the input process.
def dictCorrect(dictionary, word, definition, category):
    dictionary[word]["def"] = definition
    dictionary[word]["pos"] = category

#This statement is used to identify which words did not recieve a definition and pos from wordnet.
no_entry=[]    
for word in dictionary:
        if dictionary[word]['def'] == 'NA':
            no_entry.append(word)
            
# Words that need to be corrected: ['is', 'when', 'was', 'this']
dictCorrect(dictionary, 'is', 'third person singular present of be','v')
dictCorrect(dictionary, 'when', 'at what time','adv')
dictCorrect(dictionary, 'was', 'first and third person singular past of be','v')
dictCorrect(dictionary, 'this', 'used to identify a specific person or thing close at hand or being indicated or experienced','prn')

# Prints a polished version of the dicationary entries
for word in dictionary:
   print word +'('+dictionary[word]["pos"]+'; IPA: ['+dictionary[word]["phon"]+']): '+ dictionary[word]["def"] 



```

```python
#!/usr/bin/env python

'''Checks if two words rhyme.'''
__author__ = 'ajschwade'
__email__ = "aschwade@ucsc.edu"
__credits__ = ["Allan Jay Schwade", "Anna Greenwood"]

# This is the dictionary neeeded for the words in the test function below. 
dictionary = {}
for word in ['listen','vixen','final','file','table','label','through','rough','laughed','red']:
    dictionary[word]={
          'def':'NA',
          'pos':'NA',
          'classic_phon': 'NA', 
		  'phon_rep':{   
					  'stress_index': 999,
					  'syllables': {
                                       
					                'syll_list':[],
                                       
					                'syll0':{
					                         'phone_list':[],
					                         'nuc_index': 999,
					                 },
                                       
					                'syll1':{
					                         'phone_list':[],
					                         'nuc_index': 999,  
					                 },
                                       
					                'syll2':{
					                         'phone_list':[],
					                         'nuc_index': 999,
					                 }

					    }
				  }

       }
        

# This function allows streamlines the entry of the phonological representations need for 
# the words in the test function.
def phonEntry(dictionary, word, syllableList, syll0Phones, syll0Nuc, syll1Phones='NA', syll1Nuc='NA'):
	dictionary[word]['phon_rep']['syllables']['syll_list']=syllableList
	dictionary[word]['phon_rep']['syllables']['syll0']['phone_list']=syll0Phones
	dictionary[word]['phon_rep']['syllables']['syll0']['nuc_index']=syll0Nuc
	dictionary[word]['phon_rep']['syllables']['syll1']['phone_list']=syll1Phones
	dictionary[word]['phon_rep']['syllables']['syll1']['nuc_index']=syll1Nuc

phonEntry(dictionary, 'listen', ['lɪs','ən'], ['l','ɪ','s'], 1, ['ə','n'], 0)
phonEntry(dictionary, 'vixen', ['vɪk','sən'], ['v','ɪ','k'], 1, ['s','ə','n'], 1)
phonEntry(dictionary, 'final', ['faɪ','nəl'], ['f','aɪ'], 1, ['n','ə','l'], 1)
phonEntry(dictionary, 'file', ['faɪ','əl'], ['f','aɪ'], 1, ['ə','l'], 0)
phonEntry(dictionary, 'table', ['teɪ','bəl'], ['t','eɪ'], 1, ['b','ə','l'], 1)
phonEntry(dictionary, 'label', ['leɪ','bəl'], ['l','eɪ'], 1, ['b','ə','l'], 1)
phonEntry(dictionary, 'through', ['θɹu'], ['θ','ɹ','u'], 2)
phonEntry(dictionary, 'rough', ['ɹʌf'], ['ɹ','ʌ','f'], 1)
phonEntry(dictionary, 'laughed', ['læft'], ['l','æ','f','t'], 1)
phonEntry(dictionary, 'red', ['rɛd'], ['r','ɛ','d'], 1)


def doRhyme(w1, w2):
    """
	this function checks if two words rhyme by examining their IPA representations

    Args:
		w1: a word (string)
        w2: a word (string)
        
    Returns:
        a boolean, True if w1 and w2 rhyme, False if they do not

    Raises:
        Nothing

	"""
    # This list will hold the rhymes of the words provided
    rhyme_pair=[]
    
    # This for-loop takes two words and for each word:
    for word in [w1,w2]:
        # It finds out how many meaningful syllables it has and creates a key to access that
        # syllable's representation.
        syll_num = len(dictionary[word]['phon_rep']['syllables']['syll_list'])-1
        key='syll'+str(syll_num)
        # The index for the final syllable's nucleus is retrieved 
        nuc=dictionary[word]['phon_rep']['syllables'][key]['nuc_index']
        # Retrives the list of syllables for the word
        syllables=dictionary[word]['phon_rep']['syllables']['syll_list']
        last_syll=syllables[syll_num]
        # The rhyme is sliced from the last syllable and that string is added
        # to the rhyme pairs list .
        rhyme=last_syll[nuc:]
        rhyme_pair.append(rhyme)
    
    # If the rhymes are identical True is printed, else False.
    return True if rhyme_pair[0]==rhyme_pair[1] else False


if __name__ == "__main__":
        
    # all of the below code is used to TEST whether yours worked
    # a list of test word -> pg examples, as a dict
    wordPairsAndRhymeStatus = [ {"words": ["listen" ,"vixen" ], "rhyme": True},
                                {"words": ["final" ,"file" ], "rhyme": True},
                                {"words": ["table" ,"label" ], "rhyme": True},
                                {"words": ["through" ,"rough" ], "rhyme": False},
                                {"words": ["laughed" ,"red" ], "rhyme": False},                                
                                ]
    
    # per word, get the pg, the expected answer, and print success
    for el in wordPairsAndRhymeStatus:
        pair = el["words"]
        ans = el["rhyme"]
        rhyme = doRhyme(pair[0], pair[1])
        if rhyme == ans:
            result = "PASS"
        else:
            result = "FAIL"
        print "%10s %10s %s %s\n" % (pair[0], pair[1], rhyme, result)
```

```python
#!/usr/bin/env python


'''Creates a phonemic PigLatin generator.'''

__author__ = 'ajschwade'
__email__ = "aschwade@ucsc.edu"
__credits__ = ["Allan Jay Schwade"]

# This is the dictionary neeeded for the words in the test function below. 
dictionary = {}
for word in ['then', 'sadly', 'that', 'I', 'am', 'visit', 'actually', 'will', 'you', 'if', 'fine', 'think', 'he']:
    dictionary[word]={
          'def':'NA',
          'pos':'NA',
          'classic_phon': 'NA', 
		  'phon_rep':{   
					  'stress_index': 999,
					  'syllables': {
                                       
					                'syll_list':[],
                                       
					                'syll0':{
					                         'phone_list':[],
					                         'nuc_index': 999,
					                 },
                                       
					                'syll1':{
					                         'phone_list':[],
					                         'nuc_index': 999,  
					                 },
                                       
					                'syll2':{
					                         'phone_list':[],
					                         'nuc_index': 999,
					                 }

					    }
				  }

       }
    
 
def phonEntry(dictionary, word, pronunciation):
	dictionary[word]['classic_phon']=pronunciation
  
phonEntry(dictionary, 'then', 'ð ɛ n')
phonEntry(dictionary, 'sadly', 's æ d l i')
phonEntry(dictionary, 'that', 'ð æ t')
phonEntry(dictionary, 'I', 'aɪ')
phonEntry(dictionary, 'am', 'æ m')
phonEntry(dictionary, 'visit', 'v ɪ z ɪ t')
phonEntry(dictionary, 'actually', 'æ k tʃ ə l i')
phonEntry(dictionary, 'will', 'w ɪ l')
phonEntry(dictionary, 'you', 'j u')
phonEntry(dictionary, 'if', 'ɪ f')
phonEntry(dictionary, 'fine', 'f aɪ n')
phonEntry(dictionary, 'think', 'θ ɪ ŋ k')
phonEntry(dictionary, 'he', 'h i')


# This function is used to pig the IPA words
def pigLatin(w):
    
    """
	this function handles pigLatin for word w

    Args:
		w: a word

    Returns:
        pigLatined: a PigLatin representation of word w

    Raises:
        Nothing

	"""
    #These arrays specify which characters count as vowels and consonants in the IPA.
    vowels=['ɑ','ʌ','ɔ','aʊ','aɪ','ɛ','ɚ','eɪ','ɪ','i','oʊ','ɔɪ','ʊ','u','æ','ə']
    consonants=['b','tʃ','d','ð','f','g','h','dʒ','k','l','m','n','ŋ','ɹ','ʃ','s','t','θ','v','w','j','z','ʒ']
    
    # The input string is split into an array.
    w=w.split(' ')
    
    ### Filters ###
    # Quirky excecptions are addressed with these functions. 
    if w==['aɪ']:
        return ''.join(w)
        
    
    ### The Main Function ###
    #This for-loop looks at each character for a given word (starting with the first) and pops it to the end of the array until it finds a vowel,
    # in which case it halts and returns to resulting string plus the pig ending.
    for ch in w:
        if ch not in vowels:
            cons=w.pop(0)
            w.append(cons)
            return ''.join(w) +"ej"
        else:
            return ''.join(w) +"ej"
        

def pigLatinParts(text):
    """
	this function handles pigLatin for all words in string S. It also remembers the punctuation as at the last Goedel number.

    Args:
		text: a string

    Returns:
        pigLatined: what people can download and use (a program)
        
    Raises:
        Nothing

	"""
    
    #Defines which characters count as punctuation and creates an array to hold the pigged sentence.
    punc=[ '.' , ',' ]
    pigSentence=[]
    
    #This loop splits the input string into an array containing the words of the input sentence and pigs the words in that list 
    for word in text.split(' '):
        # If a word in the array contains some kind of puntuation, the punctuation is removed from 
        # the element, the word is pigged, and then the punctuation is re-attached. Else, the word is simply
        # pigged. The pigged array is the then printed as a string.
        if word[-1] in punc:
            hold=word[-1]
            word=word[0:-1]
            word=dictionary[word]['classic_phon']
            pigWord=pigLatin(word)+(hold)
            pigSentence.append(pigWord)
        else:
            word=dictionary[word]['classic_phon']
            pigWord=pigLatin(word)
            pigSentence.append(pigWord)
            
    return ' '.join(pigSentence)


if __name__ == "__main__":
    
    # all of the below code is used to TEST whether yours worked
    
    inputLine = "if you think, sadly, that he will actually visit, then I am fine."
 
    # I think this is right. Someone tell me if I'm wrong; there are
    # clearly some glide epentheses and ambisyllabicity issues
    # I'm avoiding here.
    
    key = "ɪfej ujej ɪŋkθej, ædlisej, ætðej ihej ɪlwej æktʃəliej ɪzɪtvej, ɛnðej aɪ æmej aɪnfej." 
    pg = pigLatinParts(inputLine)
    if pg == key:
        result = "PASS"
    else:
        result = "FAIL"
    print inputLine
    print pg
    print key
    print result
```
